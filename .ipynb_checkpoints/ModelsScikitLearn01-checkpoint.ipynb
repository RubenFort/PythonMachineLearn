# Mtricas de error
## Cmo evaluar el modelo?

<p style='text-align: justify;'>La metrica que usaremos aqui es llamada coefficiente de determinacion o simplemente $R^2$ la cual mide que tan bien una variable independiente describe la varible de respuesta. </p>

- 1 indica un ajuste perfect
- 0 indica que el modelo es equivalente a la media
- Valor negativo indica malas predicciones
- Valores de R2 aceptados rondan entre 0.7 y 0.9

<p style='text-align: justify;'>Adems de crear modelos es muy importante poder tener la habilidad para evaluarlos. Esto nos permitira tener una visin de que tan  efectivas seran las predicciones que podamos obtener.
En la pgina de [metricas de scikit-learn](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.metrics) se pueden observar diversas metricas implementadas en esta libreria. </p>

```python
from sklearn.metrics import r2_score #Importo mtica de error especfica r2_score
r2_score(y,y_hat) # y => datos verdaderos, y => estimacion
```

<p style='text-align: justify;'>El modelo de la izquierda se intenta ajustar una linea para que pase por todos los puntos de nuestro dataset. Los datos que tenemos son intrinsicamente mas complicados que una linea por lo tanto encontrar una linea que decribe nuestros datos es una tarea imposible. El modelo  nunca podra adaptarse a la perfeccin a nuestros datos por su falta de de flexibilidad para adaptarse. Se dice que el modelo tiene un bajo ajuste <font color = green>__(underfit)__</font>  a los datos. Otra forma de decir esto es que el modelo tiene un alto sesgo <font color = green>__(bias)__</font>. </p>

<p style='text-align: justify;'>El modelo de la derecha intenta ajustar un polinomio a travs de los datos.
Aqu, el ajuste del modelo tiene suficiente flexibilidad para dar cuenta casi perfectamente con todos los datos. A pesar de ello, parece que el modelo aprende las perturbaciones aleatorias de los datos. Se dice que este modelo sobreajusta <font color = green>__(overfit)__</font> el modelo por que aprendio tando de los datos como los errores aleatorios que contienen. Otra forma de decir esto es que el modelo tiene una alta varianza <font color = green>__(variance)__</font>.</p>

<p style='text-align: justify;'>Un buen modelo es aquel que puede hacer buenas predicciones con datos que no haya visto con anterioridad. Cmo podemos probar si el modelo puede predecir correctamente estos datos? </p>

## Utilizando splits

Usualmente, para detectar que tenemos un modelo adecuado se realiza una division de los datos en conjuntos de entranamiento y prueba. 
<center><img src="img/bias-variance.png" width = "60%"></center>

- __Conjunto de entrenamiento__: un subconjunto para entrenar un modelo.
- __Conjunto de pruebas__: un subconjunto para probar el modelo.

Un buena evaluacin en el conjunto de pruebas es un indicador til del buen rendimiento del modelo en general, adems es necesario tener las siguientes consideraciones:
- El conjunto de prueba debe ser lo suficientemente grande (80% del total de los datos).
- Usar diferentes conjuntos de prueba (para que el modelo no aprenda indirectamente).

<img src="img/bias-variance.svg" alt="bias-variance tradeoff" width="60%">

```python
df=_df.sample(20, random_state=100) #recuperando el dataset original
from sklearn.model_selection import train_test_split  # train_test_split => Herramienta que permite divisin de datos sencilla e intuitiva
X = df["LSTAT"].values.reshape(-1,1) # reshape(-1, 1) => (n filas desconocido, una columna)
y = df["MEDV"].values.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=100) # Devuelve 4 variables (x_entrenamiento, x_prueba, y_entrenamiento, y_prueba)
```

```python
print("df:", df.shape)
print("X:", X.shape)
print("X train:", X_train.shape)
print("X test:", X_test.shape)
```

```python
def plot(degree = 1):
    poly = PolynomialFeatures(degree=degree)
    poly.fit_transform(X_train)

    reg = LinearRegression().fit(poly.transform(X_train),y_train)

    plt.scatter(X_train,y_train)
    plt.scatter(X_test,y_test)

    y_train_hat = reg.predict(poly.transform(X_train))
    y_test_hat = reg.predict(poly.transform(X_test))

    x_model = np.linspace(0,20).reshape((-1,1))
    y_model = reg.predict(poly.transform(x_model))
    plt.plot(x_model,y_model, "r--")
    plt.axis([0, 20, 0, 40])

    print("R2 train:",r2_score(y_train,y_train_hat))
    print("R2 test:",r2_score(y_test,y_test_hat))
    
    plt.show()

interact(plot,degree=(1,10));
# interact(plot,degree=(n,m)); n => Grado del polinomio que usamos como modelo, probar valores entre 1 y 10
```

## Cross-Validation

Una alternativa al mtodo de divisin de datos, es el llamado Validacin cruzada o Cross-validation(CV). Con esta tcnica en lugar de realizar un split, se generan mltiples cortes pero en tiempos diferidos. Una forma sencilla de verlo, es imaginar que son mltiples cortes de entrenamiento  y prueba. Esta tcnica utiliza un parmetro k, el cual indica cuantas veces se repetir la divisin. Cada vez que se repite el corte se contruye un modelo de prediccin y se evalua. CV se encarga de recopilar estos resultados y presentarnos un promedio de las mtricas obtenidas. Este mtodo tiene la ventaja de utilizar todo nuestro conjunto de datos en el proceso de validacin. Por lo tanto, la mtrica de error es ms cercana al error que nos mostrara una vez puesto en produccin.


<img src="img/cv.gif" alt="CV" width="60%">

Al entrenar con un $k$=4 tendremos:

<ul>
    <li>Modelo 1 ($M_1$) validado con conjunto de prueba 1 con un error 1 ($E_1$)</li>
    <li>Modelo 2 ($M_2$) validado con conjunto de prueba 2 con un error 2 ($E_2$)</li>
    <li>Modelo 3 ($M_3$) validado con conjunto de prueba 3 con un error 3 ($E_3$)</li>
    <li>Modelo 4 ($M_4$) validado con conjunto de prueba 4 con un error 4 ($E_4$)</li>
</ul>

El error reportado por el CV es el promedio  de los $k$ ciclos utilizado.
$$E_{CV} = \frac{E_{1} + E_{2} + ... + E_{n}}{k}$$

Este mtodo tiene la ventaja de utilizar todo un conjunto de datos para entrenamiento y prueba al menos una vez. Esto nos permite no desperdiciar tantos datos como lo es cuando hacemos el test/train split.

```python
from sklearn.model_selection import KFold

#KFold(n_splits, shuffle, random_state)
kfold = KFold(4, True)

data = df[["LSTAT","MEDV"]]

for train, test in kfold.split(data):
    print('train: {}, test: {}'.format(train, test))
```

```python
from sklearn.model_selection import KFold

k = 4
kfold = KFold(k, True, 103) # KFold es quien nos da el Cross-Validation
r2_train_avg,r2_test_avg=0,0

for train, test in kfold.split(data):
    # Preparamos los datos
    X_train = data.iloc[train]["LSTAT"].values.reshape(-1,1) # Recojo la columna LSTAT de los datos de entrenamiento
    y_train = data.iloc[train]["MEDV"] # Recojo la columna MEDV de los datos de entrenamiento
    X_test = data.iloc[test]["LSTAT"].values.reshape(-1,1) # Recojo la columna LSTAT de los datos de validacion
    y_test = data.iloc[test]["MEDV"] # Recojo la columna MEDV de los datos de validacion
    # Se crea y se entrena un modelo para cada iteracin de k
    reg = LinearRegression().fit(X_train,y_train)
    # Se realizan las predicciones
    y_train_hat = reg.predict(X_train)
    y_test_hat = reg.predict(X_test)
    # Se calculan los errores
    r2_train = r2_score(y_train,y_train_hat) # r2_score => Evalua el rendimiento de un modelo de regresin lineal.
    r2_test = r2_score(y_test,y_test_hat)
    print("R2_train es {:.3f}, R2_test es {:.3f}".format(r2_train,r2_test))
    # Se acumulan  los errores para calcular el promedio
    r2_train_avg += r2_train
    r2_test_avg += r2_test
```

```python
r2_train_avg/k, r2_test_avg/k # Promedio
```

```python
from sklearn.model_selection import KFold

k= 4

kfold = KFold(k, True, 103)
r2_train_avg,r2_test_avg=0,0

for train, test in kfold.split(data):
    X_train = data.iloc[train]["LSTAT"].values.reshape(-1,1)
    y_train = data.iloc[train]["MEDV"]
    X_test = data.iloc[test]["LSTAT"].values.reshape(-1,1)
    y_test = data.iloc[test]["MEDV"]
    reg = LinearRegression().fit(X_train,y_train)
    y_train_hat = reg.predict(X_train)
    y_test_hat = reg.predict(X_test)
    r2_train = r2_score(y_train,y_train_hat)
    r2_test = r2_score(y_test,y_test_hat)
    #-----------------PLOT-------------------
    x_model = np.linspace(0,20,num=100) # 100 numeros entre 0 y 20
    y_model = reg.predict(x_model.reshape(-1,1))
    plt.scatter(X_train, y_train) # Puntos azules entrenamiento
    plt.scatter(X_test, y_test) # Puntos naranjas validacin
    plt.plot(x_model, y_model,"--r")
    plt.show()
    #----------------------------------------
    print("R2_train es {:.3f}, R2_test es {:.3f}".format(r2_train,r2_test))
    r2_train_avg += r2_train
    r2_test_avg += r2_test
```
